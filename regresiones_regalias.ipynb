```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

# Carga de datos desde resultado Consulta_regalias.ipynb

```python
Regalias = pd.read_excel('Regalias_proyectos_red_3norm.xlsx', index_col=0)

Regalias
```

```python
Regalias['IGPR']
```

```python
Regalias = Regalias[[col for col in Regalias.columns if col != 'IGPR'] + ['IGPR']]
Regalias
```

```python
Regalias.columns
```

```python
X = Regalias[[col for col in Regalias.columns if col != 'IGPR']]
Y = Regalias['IGPR']

X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=0)
```

Recordar que el diccionario de datos en la clasificación para 'Estado' es 

{'4. Desaprobado': 0, '3. Terminado': 1}

# Parámetros para Grids en GB, Regresión (Lasso y Ridge) y RF

```python
parameters_GB = {'learning_rate': np.linspace(0,1,5),
              'subsample'    : np.linspace(0,1,5),
              'n_estimators' : range(1,6),
              'max_depth'    : range(3,10)}

parameters_RF = {'max_depth':[x for x in range(2,6)], 'n_estimators':[x for x in range(1,6)]}

parameters_LR_Ridge = {'alpha': [x for x in np.linspace(0.1,10,10)]}

parameters_LR_Lasso = {'alpha': [x for x in np.linspace(0.1,10,10)]}
```

# Grid en el regresor GB 

```python
gbr = GridSearchCV(GradientBoostingRegressor(), parameters_GB, cv=5, verbose = 10)
gbr.fit(X_train, y_train)
```

```python
print('Mejor estimador: ', gbr.best_estimator_)
print('Mejor score: ',gbr.best_score_)
print('Mejores parámetros: ', gbr.best_params_)
```

# Grid en el regresor random forest

```python
rfr = GridSearchCV(RandomForestRegressor(), parameters_RF, cv=5, verbose = 10)
rfr.fit(X_train, y_train)
```

```python
print('Mejor estimador: ', rfr.best_estimator_)
print('Mejor score: ',rfr.best_score_)
print('Mejores parámetros: ', rfr.best_params_)
```

# Grid en las Regresiones lineales

Ridge

```python
lrr = GridSearchCV(Ridge(), parameters_LR_Ridge, cv=5, verbose = 10)
lrr.fit(X_train, y_train)
```

```python
print('Mejor estimador: ', lrr.best_estimator_)
print('Mejor score: ',lrr.best_score_)
print('Mejores parámetros: ', lrr.best_params_)
```

Lasso

```python
lrl = GridSearchCV(Lasso(), parameters_LR_Lasso, cv=5, verbose = 10)
lrl.fit(X_train, y_train)
```

```python
print('Mejor estimador: ', lrl.best_estimator_)
print('Mejor score: ',lrl.best_score_)
print('Mejores parámetros: ', lrl.best_params_)
```

# Regresión GB

```python
RegGB = GradientBoostingRegressor(learning_rate = gbr.best_params_.get('learning_rate'), 
                                  max_depth = gbr.best_params_.get('max_depth'), 
                                  n_estimators = gbr.best_params_.get('n_estimators'), 
                                  subsample = gbr.best_params_.get('subsample')).fit(X_train, y_train)

y_predGB = RegGB.predict(X_test)

print("R^2 : ", r2_score(y_test, y_predGB))
print("MAE :", mean_absolute_error(y_test,y_predGB))
print("RMSE:",np.sqrt(mean_squared_error(y_test, y_predGB)))
```

# Regresión Random Forest

```python
RegRF = RandomForestRegressor(max_depth = rfr.best_params_.get('max_depth'), 
                              n_estimators = rfr.best_params_.get('n_estimators')).fit(X_train, y_train)

y_predRF = RegRF.predict(X_test)

print("R^2 : ", r2_score(y_test, y_predRF))
print("MAE :", mean_absolute_error(y_test,y_predRF))
print("RMSE:",np.sqrt(mean_squared_error(y_test, y_predRF)))
```

# Regresión lineal (Normal, Lasso y Ridge)

```python
RegLR = LinearRegression().fit(X_train, y_train)

y_predLR = RegLR.predict(X_test)

print("R^2 : ", r2_score(y_test, y_predLR))
print("MAE :", mean_absolute_error(y_test,y_predLR))
print("RMSE:",np.sqrt(mean_squared_error(y_test, y_predLR)))
```

```python
RegLRR = Ridge(alpha = lrr.best_params_.get('alpha')).fit(X_train, y_train)

y_predLRR = RegLRR.predict(X_test)

print("R^2 : ", r2_score(y_test, y_predLRR))
print("MAE :", mean_absolute_error(y_test,y_predLRR))
print("RMSE:",np.sqrt(mean_squared_error(y_test, y_predLRR)))
```

```python
RegLRL = Lasso(alpha = lrr.best_params_.get('alpha')).fit(X_train, y_train)

y_predLRL = RegLRL.predict(X_test)

print("R^2 : ", r2_score(y_test, y_predLRL))
print("MAE :", mean_absolute_error(y_test,y_predLRL))
print("RMSE:",np.sqrt(mean_squared_error(y_test, y_predLRL)))
```
